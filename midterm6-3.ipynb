```python
import random
import zlib

random_string = ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=10000))

# using zlib lossless compression
compressed_data = zlib.compress(random_string.encode('utf-8'))

compression_ratio = len(compressed_data) / len(random_string.encode('utf-8'))

random_string_length = len(random_string)
compressed_data_length = len(compressed_data)

random_string_length, compressed_data_length, compression_ratio
```

```python
print("Compression Ratio: " + str(compression_ratio))
```

### Part b

The expected compression ratio in Part a is 1. This is because compression relies on observing patterns and redundancy in a string, but for a very long and TRULY random string, it should become increasingly more difficult and unlikley to compress. Each character in a truly random string has equal probability of being in any position in the string, making it hard or even impossible to compress.
